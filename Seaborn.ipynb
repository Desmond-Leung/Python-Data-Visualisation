# Scatter Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(12,4))

sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',palette='Dark2',alpha=1,s=200,
               style='level of education')

plt.savefig('My Seaborn Scatter.png')
```

## Distribution Within Categories

Box Plot:

Outliers are defined as being less than Q1 - 1.5 * IQR or more than Q3 + 1.5 * IQR.
Can have one categorical and one numerical axis and plot a boxplot.

Violin Plot:

Displays probability density across the data using a KDE.
Mirrored KDE.
Also shows the median and IQR in the plot.

Swarm Plot:

Shows all the data points in the distribution.
Can have one categorical and one numerical axis and plot a swarmplot.
WILL show how many observations in a certain category, unlike violin and boxplot.

Boxenplot (Letter-value plot):

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

## Box Plot

```python
plt.figure(figsize=(10,8),dpi=200)
sns.boxplot(data=df,x='reading score',y='parental level of education',
           hue='test preparation course',palette='Set2')

plt.legend(bbox_to_anchor=(1.05,0.5))
```

## Violin Plot

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=df,x='reading score',y='parental level of education',
              hue='test preparation course',palette='Set2',split=True,inner='quartiles')

plt.legend(bbox_to_anchor=(1.2,0.5))
```

## Swarm Plot

```python
plt.figure(figsize=(8,4),dpi=200)
sns.swarmplot(data=df,x='math score',y='gender',size=2,
             hue='test preparation course',dodge=True)

plt.legend(bbox_to_anchor=(1.25,0.5))
plt.
```

## Boxen Plot

```python
sns.boxenplot(x='math score',y='test preparation course',data=df,
             hue='gender')
```

## Joint Plots

```python
# Need two continuous features
```

```python
# Hex is useful when there are a lot of points overlapping. The hexagons are darker where there is
# more overlap. The overlap may be hard to see with the scatter kind.
```

```python
# Kinds are hist, kde, scatter and hex.
```

```python
sns.jointplot(data=df,x='math score',y='reading score',kind='hex')
```

```python
sns.jointplot(data=df,x='math score',y='reading score',kind='scatter',hue='gender')
```

## Pair Grid

```python
g = sns.PairGrid(data=df,hue='gender')

g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.histplot)

g = g.add_legend()
```
